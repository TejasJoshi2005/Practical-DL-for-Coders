```python
# import libraries

from fastai.vision.all import *
```

```python
# create the path

path = untar_data(URLs.PETS)/'images'
```

```python
# define the label function

def is_cat(x): return x[0].isupper()
```

```python
# define the data loader

dls = ImageDataLoaders.from_name_func(
    '.',get_image_files(path), valid_pct = 0.2,
    seed = 42, label_func=is_cat,
    item_tfms=Resize((192,192))
)
```

```python
# train and fine tune model

learn = vision_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(3)
```

```python
# load the weights and parameters of the model

learn.save("model_weights")
```
